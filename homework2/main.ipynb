```python
#import the pathlib and csv library
from pathlib import Path
import csv


#set the file path
csvpath = Path('../../../FinTechClass/FinTechUMiamiHomework/homework2/Documents_for_Homework_2/PyBank/Resources/budget_data.csv')

#open the csv file as an object
with open(csvpath, 'r') as csvfile:
            
    #Pass in the csv file to the reader and enable a delimiter as a comma
    csvreader = csv.reader(csvfile, delimiter=',')
           
    #Read the header row  
    header = next(csvreader)
           
    #Print the header
    #print(f"{header}")
      
    #Define variables to be used as metrics
    count_profit = 0
    total_profit = 0
    profit_by_period = []
    profit_current_period = []
    profit_prior_period = []
    change_in_profit = []
    list_of_dates = []
     
    for budget_data_row in (csvreader):
        #print(budget_data_row)
                               
        #assign variables to columns
        date = budget_data_row[0]
        profit = (int(budget_data_row[1]))   
        #print(profit)
        profit_by_period.append(profit)
        list_of_dates.append(date)
                
        #calculate metrics
        count_profit += 1
        total_profit += profit

#Define profit by periods to prep for comparison
profit_current_period = profit_by_period[1:count_profit]
profit_prior_period = profit_by_period[0:count_profit]

#Evaluate change in profit per period
zip_object = zip(profit_current_period, profit_prior_period)
for profit_current_period_i, profit_prior_period_i in zip_object:
    change_in_profit.append(profit_current_period_i - profit_prior_period_i)

#define variables to find min and max change and position in the list
maximum = max(change_in_profit) 
position_of_maximum = change_in_profit.index(maximum) + 1 #<----add 1 to get current period month
minimum = min(change_in_profit) 
position_of_minimum = change_in_profit.index(minimum) + 1 #<----add 1 to get current period month
count = len(change_in_profit)
average = round(sum(change_in_profit) / count, 2)

#print(maximum)
#print(change_in_profit)
#print(profit_prior_period)
#print(profit_current_period)
#print(change_in_profit)
#print(count)
#print(position_of_maximum)
#print(position_of_minimum)
                
#print("Final Results")
#print("______________________")
#print(" ")       
#print(f"The total of months is {count_profit}")
#(print("The total profit is " + "$" + "{:,}".format(total_profit)))
#print("The average change is "+ "$" + "{:,}".format(average))
#print("The greatest increase in profits is " + (list_of_dates[position_of_maximum]) + ": " "$" + "{:,}".format(maximum))
#print("The greatest increase in profits is " + (list_of_dates[position_of_minimum]) + ": " "$" + "{:,}".format(minimum))

# Set output file name
output_path = "final_results.txt"

#open the output file and write the results
with open(output_path, 'w') as file:
    file.write ("Final Results\n")
    file.write ("__________________\n")
    file.write (" \n")
    file.write (f"The total of months is {count_profit}\n")
    file.write ("The total profit is " + "$" + "{:,}".format(total_profit) + "\n")
    file.write ("The average change is "+ "$" + "{:,}".format(average) + "\n")
    file.write ("The greatest increase in profits is " + (list_of_dates[position_of_maximum]) + ": " "$" + "{:,}".format(maximum) + "\n")
    file.write ("The greatest increase in profits is " + (list_of_dates[position_of_minimum]) + ": " "$" + "{:,}".format(minimum) + "\n")
```
